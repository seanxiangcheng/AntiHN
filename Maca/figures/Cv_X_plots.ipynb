```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rc
rc('text', usetex=True)
from scipy import stats
import statsmodels.api as sm
```

```python
fns = ['HNNP_AFM_Cv.csv', 'HNNP_AFM_Cv_v2.csv', 'HNNP_AFM_Cv_v3.csv', 'HNNP_AFM_Cv_pre.csv']
fns = ['../data/'+fn for fn in fns]
data = {}
for fn in fns:
    data[fn] = np.loadtxt(fn, delimiter=',')
    
for fn in data:
    dt = data[fn]
    ks2plot = [16, 32, 64, 128]
    inds = np.array([x in ks2plot for x in dt[0, :]])
    
    if inds.sum()<1:
        inds = np.array([-3, -2, -1])
    muinv = dt[1:, 0]
    Ts = dt[1:, 1]
    ys = dt[1:, inds]
    plt.figure(figsize=(5,5))
    plt.plot(muinv, ys)
```
